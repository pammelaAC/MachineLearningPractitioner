```python
import cv2
```

```python
model = YOLO("/content/sample_data/best.pt")
```

```python
img_path = "/content/friends.jpg"
```

```python
results = model(img_path)
```

```python
boxes = results[0].boxes
```

```python
img = cv2.imread(img_path)
```

```python
for box in boxes:
  top_left_x = int(box.xyxy.tolist()[0][0])
  top_left_y = int(box.xyxy.tolist()[0][1])
  bottom_right_x = int(box.xyxy.tolist()[0][2])
  bottom_right_y = int(box.xyxy.tolist()[0][3])

  cv2.rectangle(img,(top_left_x,top_left_y),(bottom_right_x,bottom_right_y),(50,200,129),2)

  cv2.imwrite("testing.jpeg", img)
```

```python
img = mpimg.imread('testing.jpeg')
plt.imshow(img)
plt.axis('off')
plt.show()
```
